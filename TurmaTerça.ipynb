# Seleção baseado em rótulo

O paradigma para seleção de atributos é o operador __loc__: _seleção baseada em rótulos_. 

Nesse paradigma, filtramos o __valor__ das linhas pelos seus índices e as colunas pelo seu __nome__.

Por exemplo, para obter a primeira entrada do _dataset_, agora faríamos o seguinte: `df.loc[0, 'coluna']`

Podemos também fazer uma busca usando uma lista de colunas para quando quisermos pegar mais de uma usando o __loc__: `df.loc[:, ['coluna', 'coluna', 'coluna']]`

````python
#Lendo a primeira linhas da coluna 'região'
df.loc[0, 'regiao']
````

````python
#Pegando todas as linhas e pegando as primerias 4 colunas da tabela 
df.loc[ : , ['regiao',	'estado', 'municipio', 'coduf']]
````

````python
# Filtrando as primeiras 10 linhas da coluna região
df.loc[0:10,'regiao']
````

````python
#Transformando consulta em dataframe
df.loc[0:10, 'regiao'].to_frame()
````

````python
#Exibindo as primeiras 10 linhas de todas as colunas
df.loc[0:10,:]
````

````python
#Se deixarmos ambos os campos somente com os dois pontos temos o seguinte
df.loc[:,:]
````

````python
#Pegando um intervalo de colunas com o loc
df.loc[0:10,'regiao':'semanaEpi']
````

# Seleção baseado em índice -> iloc

Se por um lado o __loc__ localiza os elementos para nós baseado nos nomes das colunas e nos index o __iloc__ só irá conseguir localizar os index de nossas linhas e das nossas colunas.

__SINTAXE__

`df.iloc[0,0]`

````python
df.iloc[0]
````

````python
df.iloc[0,0]
````

````python
df.iloc[0].to_frame()
````

````python
df.iloc[0].to_frame().T
````


Tanto loc quanto iloc trabalham com linha-primeiro e coluna-depois.

Isto significa que é um pouco mais fácil recuperar linhas e um pouco mais difícil recuperar colunas. Para obter uma coluna com iloc, devemos escrever assim: df.iloc[:, 0]

Por si só, o seletor ' : ' , que também vem do Python nativo, significa tudo.

Quando combinado com outros seletores, no entanto, pode ser usado para indicar um intervalo de valores.

Por exemplo, para selecionar a coluna da Data da Venda apenas da primeira, segunda e terceira linha, devemos seguir a sintaxe df.iloc[:3, 0], mas, também, podemos passar uma lista como df.iloc[[0, 1, 2, 3], 0].

Podemos passar um intervalo, por exemplo: df.iloc[1:3, 0]

Por fim, vale saber que números negativos podem ser usados na seleção. Isso começará a contar para frente a partir do final dos valores.

Então, por exemplo, aqui estão os últimos cinco elementos do conjunto de dados - df.iloc[-5:]


````python
df.iloc[-5:,]
````

# Groupby

Provavelmente essa palavra já é conhecida para quem usa __SQL__, pois o conceito do group by é que basicamente possamos agrupar os nossos dados em grupos

````python
data = {'turmas':['1A','1A','2A','2A','3A','3A','3A'],
        'alunos': ['Carlos','Grazi','Nator', 'Juci', 'Miguel', 'Aline', 'Emanuel'],
        'notas':[7,8,10,6,5,4.5,10]
        }
        
escola = pd.DataFrame(data)
escola
````

````python
escola['notas'].mean()
````

Vamos supor que desejamos agrupar os nossos dados por turma para saber a média das turmas, mas como poderiamos fazer isso? Poderiamos fazer isso dá seguinte maneira: 

````python
#Quando utilizo esse comando ele irá me gerar um objeto da classe groupby e devolvera o endereçamento na memória do computador
escola.groupby("turmas")
````

Mas como podemos deixar isso mais útil? Bem podemos usar as funções de agregação para dizer para ele o que desejamos fazer com esse agrupamento de dados. Por exemplo vamos ver quantos alunos e notas existem no:  

*   1A
*   2B
*   3B

Eu posso me utilizar das funções de agregação -- Uma função de agregação processa um conjunto de valores contidos em uma única coluna de uma tabela e retorna um único valor como resultado.-- para saber a quantidade de alunos em cada turma, a somatoria de notas de cada turma e até mesmo a média de notas de cada __turma__

````python
escola.groupby('turmas').count()
````

````python
escola.groupby("turmas").mean()
````

As funções de agregação que mais utilizamos são: 

1.   __sum()__
2.   __mean()__
3.   __max()__
4.   __min()__
5.   __median()__ 
6.   __count()__





Posso usar o groupby sem especificar nenhuma das minhas colunas, porém o __'problema'__ de se fazer isso é que ele irá retornar todas as colunas com a agregação que utilizamos

````python
paises = pd.DataFrame({
    "country" : ["United States", "Canada", "United Kingdom", "France", "Germany", "China", "Japan", "South Korea"],
    "continent" : ["North America", "North America", "Europe", "Europe", "Europe", "Asia", "Asia", "Asia"],
    "population" : [332722557, 38711108, 67081234, 67853000, 83222442, 1412600000, 125502000, 51745000],
    "area" : [3796742, 3855100, 93628, 247368, 137882, 3705407, 145937, 38690],
    "population percentage of world": [4.18, 0.487, 0.843, 0.853, 1.05, 17.8, 1.58, 0.650]
})

paises
````

````python
paises.groupby('continent').sum()
````

# Gráficos

Como vimos anteriormente posso fazer gráficos usando a biblioteca plotly que já vem por pandrão no collab ou no anaconda.

Aqui iremos explorar um pouco dessa biblioteca com a seguinte base de dados.

`https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv`

Essa base de dados registra o número de gorjetas recebidas em alguns meses em um restaurante.

*   total bill -> conta
*   Tip -> gorjeta
*   sex -> sexo
*   Smoker -> fumante
*   Item da lista
*   day -> dia da semana
*   time -> hora do dia 
*   size -> número de pessoas 






````python
df_conta = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df_conta
````

````python
df_conta_agrupamento_dia = df_conta.groupby('day').agg({'total_bill': sum}).reset_index()

df_conta_agrupamento_dia
````

````python
px.bar(df_conta_agrupamento_dia, x='day', y='total_bill')
````

````python
df_conta_agrupamento_dia_sexo = df_conta.groupby(['day', 'sex'], sort=True).agg({'total_bill': sum}).reset_index()
df_conta_agrupamento_dia_sexo
````

# Gráfico customizado
Podemos customizar os gráficos de 'n' maneiras veremos alguns deles a seguir 



*   `px.bar(df, x='column')` -> Irá definir o eixo X da nossa coluna
*   `px.bar(df, x='column', y='column')` -> Irá Definir o eixo Y
*   `px.bar(df, x='column', y='column', color='column')` -> Irá definir uma cor para cada tipo de dado de uma coluna
*   `px.bar(df, x='column', y='column', color='column', barmode='group')` -> Irá 



````python
px.bar(df_conta_agrupamento_dia_sexo, x='day', y='total_bill', color = 'sex', 'barmode='group'')
````
